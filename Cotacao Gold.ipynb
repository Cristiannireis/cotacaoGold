```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import matplotlib.pyplot as plt

cotacaoGold= web.DataReader('GC=F', data_source='yahoo', start='2022-07-03', end='2022-08-03')
print(cotacaoGold)
```

```python
cotacaoGold['Adj Close'].plot(figsize=(15,5))
```

```python
retorno_Gold = cotacaoGold['Adj Close'][-1] / cotacaoGold['Adj Close'][0]-1
print('Retorno do Gold {:.2%}'.format(retorno_Gold))
```

```python
cotacaoGold['Adj Close'].plot(figsize=(15,5), label = 'cotacaoGold')
cotacaoGold['Adj Close'].rolling(3).mean().plot(label = 'MM 3')
cotacaoGold['Adj Close'].rolling(5).mean().plot(label = 'MM 5')
plt.legend()
plt.show()
```
